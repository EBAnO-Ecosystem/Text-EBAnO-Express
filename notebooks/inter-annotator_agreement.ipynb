```python
import simpledorff
import pandas as pd
df_full = pd.read_csv('T-EBANO Survey.csv') #Load Your Dataframe
df_full.head()
```

```python
df = df_full.iloc[:,3:-3]
```

```python
df.columns
```

```python
for row_id, row in df.iterrows():
    for index, value in row.items():
        print(f"Index : {index}, Value : {value}")
```

```python
def get_annotation(v):
    if v == "Not relevant":
        return "A"
    elif v == "Relevant":
        return "B"
    else:
        return "B"
```

```python
annotator_id = 0
accumulator = []
for row_id, row in df.iterrows():
    document_id = 0
    for index, value in row.items():
        annotation = get_annotation(value)
        accumulator.append({"document_id":document_id, "annotator_id":annotator_id, "annotation":annotation})
        document_id += 1
    annotator_id += 1
```

```python
df_converted = pd.DataFrame(accumulator)
```
